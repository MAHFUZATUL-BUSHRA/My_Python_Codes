```python
# see the loaded image in Colab
import PIL
from PIL import ImageDraw
img = PIL.Image.open("receipt.png")
img
```

```python
# draw the bounding boxes
import easyocr

# this needs to run only once to load the model into memory
en_reader = easyocr.Reader(['en'])

bounding_box = en_reader.readtext('/content/receipt.png')
bounding_box
```

```python
# Draw bounding boxes

import PIL

def draw_boxes(image, bounding_points, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounding_points:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image
```

```python
draw_boxes(img, bounding_box)
img
```

# EasyOCR on BanglaText

```python
import PIL
from PIL import ImageDraw
import easyocr
import cv2

# this needs to run only once to load the model into memory
bn_reader = easyocr.Reader(['bn'])

def draw_boxes(image, bounding_points, color='green', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounding_points:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

# read the photo
bn_img = PIL.Image.open('bangla.jpg')
bounding_box = bn_reader.readtext(bn_img)

# draw the bounding boxes
draw_boxes(bn_img, bounding_box)

bn_img
```
